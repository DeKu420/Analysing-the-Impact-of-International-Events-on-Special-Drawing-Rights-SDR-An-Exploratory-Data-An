```python
import pandas as pd
import spacy
# Load spaCy model
nlp = spacy.load("en_core_web_sm")
```

```python
countries
```

```python
def convert_to_geographical(country):
    if country not in geopolitical_to_geographical_mapping.keys():
        return country
    return geopolitical_to_geographical_mapping[country]

# Function to extract country from the event
def extract_country(event):
    doc = nlp(event)
    for ent in doc.ents:
        if ent.label_ == "GPE" or ent.label_ =="ORG" or ent.label_ == "NORP" or ent.label_ == "PERSON":  # GPE represents geopolitical entity
            country = ent.text
            return convert_to_geographical(country)
    return convert_to_geographical(doc.text.split(" ")[0])  # If no country found, mark as NIL
```

```python
extract_country("Omani")
```

```python
countries = [extract_country(column_name) for column_name in countries]
```

```python
data.columns = ['Data']+countries
```
